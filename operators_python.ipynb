# operators in python

## 6 types of operators in python
1. arithametic operators
2. assignment operators
3. comparison operators
4. logical operators
5. membership operators
6. identity operators

```python
# arithametic operators - +,-./,//,%
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
#assignment operators - +=,-=,*=,/=,//=,%=
a=100
a+=50
print(a)
```
